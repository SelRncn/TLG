***4.	Analizando la información proporcionada, detectaste que no se cuenta con los años que el empelado lleva en la compañía y parece ser un buen dato. Dicha cantidad se puede calcular con la fecha de contratación ‘HiringDate’:***
*   Crea una columna llamada Year y obtén el año de contratación del empleado a partir de su fecha ‘HiringDate’. No se te olvide que debe ser un entero.
*   Crea una columna llamada YearsAtCompany que contenga los años que el empleado lleva en la compañía hasta el año 2018. Para su cálculo, usa la variable Year que acabas de crear.



```python
# Crear la columna "Year" obteniendo el año de la fecha de contratación
EmpleadosAttrition['Year'] = EmpleadosAttrition['HiringDate'].str.split('/', expand=True)[2]
EmpleadosAttrition['Year'] = pd.to_numeric(EmpleadosAttrition['Year']).astype('Int64')

# Calcular los años en la empresa hasta el 2018 "YearsAtCompany"
EmpleadosAttrition['YearsAtCompany'] = 2018 - EmpleadosAttrition['Year']
EmpleadosAttrition.head()
```

***9. Convierte todas las variables categóricas que quedan a numéricas:***

a.	BusinessTravel
b.	Department
c.	EducationField
d.	Gender
e.	JobRole
f.	MaritalStatus

```python
Cathegories = ['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus','Attrition','OverTime']
EmpleadosAttrition_OH = pd.get_dummies(EmpleadosAttrition, columns= Cathegories, dtype= np.int64)
EmpleadosAttrition_OH.head()
```

```python
# Conocer los nuevos índices de las columnas que se crearon
EmpleadosAttrition_OH.info()
```

```python
# Eliminar columnas redundantes, en las de tipo clase binaria y/o convertir a una misma columna
# con valores de 1 (Attrition, Female y OverTime) y 0 (Not Atrition, Male, No OverTime)
EmpleadosAttrition_OH.drop(columns=['Attrition_No','Gender_Male','OverTime_No'])
```

***10. Calcula la correlación lineal de cada una de las variables con respecto al Attrition***

```python
Correlacion = EmpleadosAttrition_OH.corr()['Attrition_Yes'].drop('Attrition_Yes')
display('Correlación de cada variable de entrada respecto a la salida: \n', Correlacion)
```

***11.	Selecciona solo aquellas variables que tengan una correlación mayor o igual a 0.1, dejándolas en otro frame llamado EmpleadosAttritionFinal***

No olvides mantener la variable de salida Attrition; esto es equivalente a borrar las que no cumplen con el límite.

```python
# Filtrar las variables que cumplen nuestro requisito
VariablesCorrelacion = Correlacion[Correlacion.abs() > 0.1].index

# Usar las columnas que filtradas en el paso anterior
EmpleadosAttritionFinal = EmpleadosAttrition_OH[VariablesCorrelacion.tolist() + ['Attrition_Yes']]
EmpleadosAttritionFinal.head()
```

***12.	Crea una nueva variable llamada EmpleadosAttritionPCA formada por los componentes principales del frame EmpleadosAttritionFinal***

Recuerda que el resultado del proceso PCA es un numpy array, por lo que, para hacer referencia a una columna, por ejemplo, la 0, puedes usar la instrucción EmpleadosAttritionPCA[:,0]).


```python
Pca = PCA()
Pca.fit(EmpleadosAttritionFinal)

display(Pca.components_)
```

```python
# Valoración de los componentes de PCA
print(Pca.explained_variance_)
```

```python
# Valoración del %Componentes de PCA
print(Pca.explained_variance_ratio_)
```

```python
# Aplicar componentes del PCA al DF EmpleadosAttritionFinal
EmpleadosAttritionPCA = Pca.transform(EmpleadosAttritionFinal)
EmpleadosAttritionPCA
```

***13. Agrega el mínimo número de Componentes Principales en columnas del frame EmpleadosAttritionPCA que logren explicar el 80% de la varianza, al frame EmpleadosAttritionFinal***

Puedes usar la instrucción assign, columna por columna, llamando a cada una C0, C1, etc., hasta las que vayas a agregar.

```python
# Obteniendo los valores mínimos para obtener el 80% de varianza
SumatoriaVarianza = np.cumsum(Pca.explained_variance_ratio_)
n_Pca = np.argmax(SumatoriaVarianza >= 0.8) + 1

# Seleccionar los primeros valores PCA
Pca_Top = EmpleadosAttritionPCA[:, :n_Pca]
```

```python
# Nombrar los valores
Nombres_Pca = [f'C{i}' for i in range(n_Pca)]

# Convertir los valores a un DF
Pca_df = pd.DataFrame(Pca_Top, columns= Nombres_Pca)
Pca_df.head()
```

```python
# Añadir valores al DF original EmpleadosAtrittionFinal
EmpleadosAttritionFinal = EmpleadosAttritionFinal.assign(**Pca_df)
EmpleadosAttritionFinal.head()
```
